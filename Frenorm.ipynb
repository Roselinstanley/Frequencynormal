```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("prekidney.csv")
quan=[]
qual=[]
dataset
```

```python
df=pd.DataFrame(dataset)
dfid=pd.DataFrame(dataset['id'])
```

```python
df=df.drop(["id"],axis=1)
dataset=df
```

```python
#Splitting numerical and categorical data
from Univariatekidney import univar
u=univar(dataset)
qual,quan=u.quanqual()
```

```python
# In the dataset PCV,RC,WC are  umeric values but its typr is object so it is been converted to numeric value
dataset['pcv']=pd.to_numeric(dataset['pcv'],errors='coerce')
dataset['wc']=pd.to_numeric(dataset['wc'],errors='coerce')
dataset['rc']=pd.to_numeric(dataset['rc'],errors='coerce')
```

```python
dataset.isnull().sum()
```

```python
from Freqnorm import Freqnorm
f=Freqnorm(dataset)
for cname in quan:   
    freq,skewness,kurtosis,variance,std=f.frequency(cname)     
    print("\n\n",cname,"\n\n")
    print(freq)
    print("\n\nskewness value of",cname,"=",skewness)
    print("\n\nKurtosis value of",cname,"=",kurtosis)
    print("\n\nVariance value of",cname,"=",variance)
    print("\n\nSD value of",cname,"=",std)
```
